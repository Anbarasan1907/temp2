```python
import pandas as pd
import matplotlib.pyplot as plt

dict={'Age':[10,20,30,40,50],'Salary':[25000,35000,45000,55000,65000]}
data=pd.DataFrame(dict)
print(data)
print(data.head(2))
print(data.tail(2))
print(data.info())
print(data.describe())
print(data.shape)
plt.plot(data['Age'], data['Salary'])

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("Age vs Salary")

plt.show()
```
